```python
import pandas as pd
import xml.etree.ElementTree as ET
from pathlib import Path
from google.colab import files



def convert_csv_to_excel(input_file, output_file):
    df = pd.read_csv(input_file)
    df.to_excel(output_file, index=False)

def convert_excel_to_csv(input_file, output_file):
    df = pd.read_excel(input_file, engine='openpyxl')
    df.to_csv(output_file, index=False)

def convert_csv_to_xml(input_file, output_file):
    df = pd.read_csv(input_file)
    root = ET.Element("root")
    for _, row in df.iterrows():
        item = ET.SubElement(root, "row")
        for col in df.columns:
            child = ET.SubElement(item, col)
            child.text = str(row[col])
    tree = ET.ElementTree(root)
    tree.write(output_file, encoding="utf-8", xml_declaration=True)

def convert_xml_to_csv(input_file, output_file):
    tree = ET.parse(input_file)
    root = tree.getroot()
    data = []
    for row in root.findall("row"):
        data.append({child.tag: child.text for child in row})
    df = pd.DataFrame(data)
    df.to_csv(output_file, index=False)

def convert_excel_to_xml(input_file, output_file):
    df = pd.read_excel(input_file, engine='openpyxl')
    root = ET.Element("root")
    for _, row in df.iterrows():
        item = ET.SubElement(root, "row")
        for col in df.columns:
            child = ET.SubElement(item, col)
            child.text = str(row[col])
    tree = ET.ElementTree(root)
    tree.write(output_file, encoding="utf-8", xml_declaration=True)

def convert_xml_to_excel(input_file, output_file):
    tree = ET.parse(input_file)
    root = tree.getroot()
    data = []
    for row in root.findall("row"):
        data.append({child.tag: child.text for child in row})
    df = pd.DataFrame(data)
    df.to_excel(output_file, index=False)



def convert_file(input_file, target_format):
    input_path = Path(input_file)
    input_ext = input_path.suffix.lower()
    output_file = input_path.stem + f".{target_format}"

    if input_ext == ".csv":
        if target_format == "xlsx":
            convert_csv_to_excel(input_file, output_file)
        elif target_format == "xml":
            convert_csv_to_xml(input_file, output_file)
        else:
            print(" Unsupported target format for CSV.")
            return None

    elif input_ext == ".xlsx":
        if target_format == "csv":
            convert_excel_to_csv(input_file, output_file)
        elif target_format == "xml":
            convert_excel_to_xml(input_file, output_file)
        else:
            print(" Unsupported target format for Excel.")
            return None

    elif input_ext == ".xml":
        if target_format == "csv":
            convert_xml_to_csv(input_file, output_file)
        elif target_format == "xlsx":
            convert_xml_to_excel(input_file, output_file)
        else:
            print(" Unsupported target format for XML.")
            return None

    else:
        print(" Unsupported source file type.")
        return None

    print(f" Converted to: {output_file}")
```

```python
import pandas as pd
import xml.etree.ElementTree as ET
from pathlib import Path
from google.colab import files
import tabula
from fpdf import FPDF
import xml.dom.minidom as minidom



def prettify_xml(elem):
    rough_string = ET.tostring(elem, encoding="utf-8")
    reparsed = minidom.parseString(rough_string)
    return reparsed.toprettyxml(indent="  ")



def convert_csv_to_excel(input_file, output_file):
    df = pd.read_csv(input_file)
    df.to_excel(output_file, index=False)

def convert_excel_to_csv(input_file, output_file):
    df = pd.read_excel(input_file, engine='openpyxl')
    df.to_csv(output_file, index=False)



def convert_csv_to_xml(input_file, output_file):
    df = pd.read_csv(input_file)
    root = ET.Element("root")
    for _, row in df.iterrows():
        item = ET.SubElement(root, "row")
        for col in df.columns:
            child = ET.SubElement(item, col)
            child.text = "" if pd.isna(row[col]) else str(row[col])
    with open(output_file, "w", encoding="utf-8") as f:
        f.write(prettify_xml(root))

def convert_xml_to_csv(input_file, output_file):
    tree = ET.parse(input_file)
    root = tree.getroot()
    data = []
    for row in root.findall("row"):
        data.append({child.tag: child.text for child in row})
    df = pd.DataFrame(data)
    df.to_csv(output_file, index=False)



def convert_excel_to_xml(input_file, output_file):
    df = pd.read_excel(input_file, engine='openpyxl')
    root = ET.Element("root")
    for _, row in df.iterrows():
        item = ET.SubElement(root, "row")
        for col in df.columns:
            child = ET.SubElement(item, col)
            child.text = "" if pd.isna(row[col]) else str(row[col])
    with open(output_file, "w", encoding="utf-8") as f:
        f.write(prettify_xml(root))

def convert_xml_to_excel(input_file, output_file):
    tree = ET.parse(input_file)
    root = tree.getroot()
    data = []
    for row in root.findall("row"):
        data.append({child.tag: child.text for child in row})
    df = pd.DataFrame(data)
    df.to_excel(output_file, index=False)



# PDF → CSV/Excel/XML
def convert_pdf_to_csv(input_file, output_file):
    dfs = tabula.read_pdf(input_file, pages='all', multiple_tables=True)
    if dfs:
        combined_df = pd.concat(dfs, ignore_index=True)
        combined_df.to_csv(output_file, index=False)

def convert_pdf_to_excel(input_file, output_file):
    dfs = tabula.read_pdf(input_file, pages='all', multiple_tables=True)
    if dfs:
        combined_df = pd.concat(dfs, ignore_index=True)
        combined_df.to_excel(output_file, index=False)

def convert_pdf_to_xml(input_file, output_file):
    dfs = tabula.read_pdf(input_file, pages='all', multiple_tables=True)
    if dfs:
        combined_df = pd.concat(dfs, ignore_index=True)
        root = ET.Element("root")
        for _, row in combined_df.iterrows():
            item = ET.SubElement(root, "row")
            for col in combined_df.columns:
                child = ET.SubElement(item, str(col))
                child.text = "" if pd.isna(row[col]) else str(row[col])
        with open(output_file, "w", encoding="utf-8") as f:
            f.write(prettify_xml(root))

# CSV/Excel/XML → PDF
def dataframe_to_pdf(df, output_file):
    pdf = FPDF()
    pdf.set_font("Arial", size=10)
    pdf.add_page()

    # Table header
    col_width = pdf.w / (len(df.columns) + 1)
    for col in df.columns:
        pdf.cell(col_width, 8, str(col), border=1)
    pdf.ln()

    # Table rows
    for _, row in df.iterrows():
        for value in row:
            pdf.cell(col_width, 8, str(value), border=1)
        pdf.ln()

    pdf.output(output_file)

def convert_csv_to_pdf(input_file, output_file):
    df = pd.read_csv(input_file)
    dataframe_to_pdf(df, output_file)

def convert_excel_to_pdf(input_file, output_file):
    df = pd.read_excel(input_file, engine='openpyxl')
    dataframe_to_pdf(df, output_file)

def convert_xml_to_pdf(input_file, output_file):
    tree = ET.parse(input_file)
    root = tree.getroot()
    data = []
    for row in root.findall("row"):
        data.append({child.tag: child.text for child in row})
    df = pd.DataFrame(data)
    dataframe_to_pdf(df, output_file)



def convert_file(input_file, target_format):
    input_path = Path(input_file)
    input_ext = input_path.suffix.lower()
    output_file = input_path.stem + f".{target_format}"

    try:
        if input_ext == ".csv":
            if target_format == "xlsx":
                convert_csv_to_excel(input_file, output_file)
            elif target_format == "xml":
                convert_csv_to_xml(input_file, output_file)
            elif target_format == "pdf":
                convert_csv_to_pdf(input_file, output_file)
            else:
                print("Unsupported target format for CSV.")
                return None

        elif input_ext == ".xlsx":
            if target_format == "csv":
                convert_excel_to_csv(input_file, output_file)
            elif target_format == "xml":
                convert_excel_to_xml(input_file, output_file)
            elif target_format == "pdf":
                convert_excel_to_pdf(input_file, output_file)
            else:
                print("Unsupported target format for Excel.")
                return None

        elif input_ext == ".xml":
            if target_format == "csv":
                convert_xml_to_csv(input_file, output_file)
            elif target_format == "xlsx":
                convert_xml_to_excel(input_file, output_file)
            elif target_format == "pdf":
                convert_xml_to_pdf(input_file, output_file)
            else:
                print("Unsupported target format for XML.")
                return None

        elif input_ext == ".pdf":
            if target_format == "csv":
                convert_pdf_to_csv(input_file, output_file)
            elif target_format == "xlsx":
                convert_pdf_to_excel(input_file, output_file)
            elif target_format == "xml":
                convert_pdf_to_xml(input_file, output_file)
            else:
                print("Unsupported target format for PDF.")
                return None

        else:
            print("Unsupported source file type.")
            return None

        print(f"Converted to: {output_file}")
    except Exception as e:
        print(f"Error: {e}")
```
